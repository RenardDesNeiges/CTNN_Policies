# Visualizing neuron's behaviors

### Import requirements

```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
def activation_amplitude(amplitude):
    dt = 0.001
    t = np.arange(0,20,dt)
    def LTC(f,tau,activation,A):
        return x + dt -(1/tau + activation)*x - A*activation

    x0 = 20
    x = x0
    xs = []

    activation = np.sin(np.arange(0,20,dt)*0.001*(1/dt))*0.001*amplitude
    # activation = np.zeros(t.shape)
    tau = 1000
    A = 0

    for i in range(t.shape[0]):
        x = LTC(x,tau,activation[i],A)
        xs.append(x)
    np.array(xs)

    plt.plot(t,xs)
    plt.plot(t,activation)
    plt.show()
    
interact(activation_amplitude, amplitude=10.)
```
